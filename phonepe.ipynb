```python
import os
import json
import pandas as pd
import psycopg2
```

```python
# aggregated_transcation

country_path1 = "C:/Users/razih/Desktop/New folder/.venv/phonepe/pulse/data/aggregated/transaction/country/india/state/"

ag_transaction_list =os.listdir(country_path1)

colmns1={"states":[],"years":[], "quarter":[], "Trans_type":[],"Trans_count":[], "Trans_amount":[]}


for st in ag_transaction_list:
    current_states = country_path1 + st + "/"
    #print(current_states)
    list_of_years = os.listdir(current_states)
    #print(list_of_years)
    
    for yr in list_of_years:
        current_year= current_states + yr + "/"
        #print(current_year)
        list_of_files=os.listdir(current_year)
        #print(list_of_files)
        for fi in list_of_files:
            current_file= current_year + fi
            #print(current_file)
            aggr_data=open(current_file, "r")
            #print(aggr_data)

            A_t_data=json.load(aggr_data)
            for i in A_t_data['data']['transactionData']:
                name=i['name']
                #print(name)      
                count=i['paymentInstruments'][0]['count']
                #print(count)
                amount=i['paymentInstruments'][0]['amount']
                #print(amount)

                colmns1['Trans_type'].append(name)
                colmns1['Trans_count'].append(count)
                colmns1['Trans_amount'].append(amount)
                colmns1['states'].append(st)
                colmns1['years'].append(yr)
                colmns1['quarter'].append(int(fi.strip(".json")))

                #print(colmns1)

                aggregated_transaction = pd.DataFrame(colmns1)
                
                aggregated_transaction["states"]=aggregated_transaction["states"].str.replace("andaman-&-nicobar-islands","andaman & nicobar")
                aggregated_transaction["states"]=aggregated_transaction["states"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","dadra and nagar-haveli-and-daman-and-diu")
                aggregated_transaction["states"]=aggregated_transaction["states"].str.replace("-"," ")
                aggregated_transaction["states"]=aggregated_transaction["states"].str.title()
```

```python
# aggregated user
user_path1= "C:/Users/razih/Desktop/New folder/.venv/phonepe/pulse/data/aggregated/user/country/india/state/"

ag_users_list =os.listdir(user_path1)

colmns2={"states":[],"years":[], "quarter":[], "Brand":[],"Trans_count":[], "Percentage":[]}


for st in ag_transaction_list:
    current_states = user_path1 + st + "/"
    #print(current_states)

    list_of_years = os.listdir(current_states)
    #print(list_of_years)
    
    for yr in list_of_years:
        current_year= current_states + yr + "/"
        #print(current_year)
        
        list_of_files=os.listdir(current_year)
        #print(list_of_files)
        for fi in list_of_files:
            current_file= current_year + fi
            #print(current_file)
            aggr_data=open(current_file, "r")
            #print(aggr_data)

            A_U_data=json.load(aggr_data)
            #print(A_U_data)
            try:
                for i in A_U_data['data']['usersByDevice']:
                    brand=i['brand']
                    #print(brand)      
                    count=i['count']
                    #print(count)
                    percentage=i['percentage']
                    #print(percentage)

                    colmns2['Brand'].append(brand)
                    colmns2['Trans_count'].append(count)
                    colmns2['Percentage'].append(percentage)
                    colmns2['states'].append(st)
                    colmns2['years'].append(yr)
                    colmns2['quarter'].append(int(fi.strip(".json")))
            except:
                pass
                
                aggregated_users= pd.DataFrame(colmns2)
                
                aggregated_users["states"]=aggregated_users["states"].str.replace("andaman-&-nicobar-islands","andaman & nicobar")
                aggregated_users["states"]=aggregated_users["states"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","dadra and nagar-haveli-and-daman-and-diu")
                aggregated_users["states"]=aggregated_users["states"].str.replace("-"," ")

                aggregated_users["states"]=aggregated_users["states"].str.title()   
```
